```python
import os
```

```python
file=['twitchDE', 'twitchENGB', 'twitchES', 'twitchFR', 'twitchPTBR', 'twitchRU']
contagions=['SimpleContagion', 'ComplexContagion']
startRatio=['01', '02', '03', '04', '05']
for f in file:
    for contagion in contagions:
        for sRatio in startRatio:
            folder='./'+f+'/'+contagion+'/'+sRatio+'/'
            for filename in os.listdir(folder):
                print(filename)
                nfilename=filename.replace("_", "", 1)
                os.rename(folder+filename, folder+nfilename)
```
