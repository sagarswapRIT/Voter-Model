```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
output_folder_path = "../../data/output/RealWorld/"
input_folder_path="../../data/input/RealWorld/"
datasets=['astroPh', 'emailEU', 'facebook', 'facebookMedium', 'facebookArtist', 'gitInteractions', 'hepPh', 'lastfm', 'redditHyperlink', 'twitch_DE', 'twitch_ENGB', 'twitch_ES', 'twitch_FR', 'twitch_PTBR', 'twitch_RU']
contagions=['SimpleContagion', 'ComplexContagion']
startRatio=['01', '02', '03', '04', '05']
```

```python
def create_graphs(folder, dataset, contagion, sRatio):
    files=os.listdir(folder)
    data0=[]
    data10=[]
    data20=[]
    data30=[]
    data40=[]
    data50=[]
    data60=[]
    data70=[]
    data80=[]
    data90=[]
    data100=[]

    for file in files:
        with open(os.path.join(folder, file), 'r') as f:
            content = pd.read_csv(folder+"/"+file, names=['Epoch', 'Pop', 'Frac', 'DiscEdge'], skiprows=1, sep=" ")
            file=file.split('_')

            content=content.drop(['Epoch', 'Pop'], axis=1)    
            last_row_content=content.iloc[-1, 0]
            if last_row_content>0.5:
                last_row_content=1-last_row_content
            if file[1]=='r0':
                data0.append(float(last_row_content))
            elif file[1]=='r10':
                data10.append(float(last_row_content))
            elif file[1]=='r20':
                data20.append(float(last_row_content))
            elif file[1]=='r30':
                data30.append(float(last_row_content))
            elif file[1]=='r40':
                data40.append(float(last_row_content))
            elif file[1]=='r50':
                data50.append(float(last_row_content))
            elif file[1]=='r60':
                data60.append(float(last_row_content))
            elif file[1]=='r70':
                data70.append(float(last_row_content))
            elif file[1]=='r80':
                data80.append(float(last_row_content))
            elif file[1]=='r90':
                data90.append(float(last_row_content))
            elif file[1]=='r100':
                data100.append(float(last_row_content))

    my_dict={'0.0':data0, '0.1':data10, '0.2':data20, '0.3':data30, '0.4':data40, '0.5':data50, '0.6':data60, '0.7':data70, '0.8':data80, '0.9':data90, '1.0':data100}
    #print(my_dict)
    fig, ax = plt.subplots()
    ax.boxplot(my_dict.values())
    ax.set_xticklabels(my_dict.keys())
    ax.set_xlabel("Rewiring Rate ($\\alpha$)")
    ax.set_ylabel("MinorityOpinion Density ($\\rho$)")
    plt.title("Minority Opinion Density vs Rewiring Rate for "+dataset+" with "+contagion)
    plt.show()
    figpath="../../figures/RealWorld/"+dataset+"/"+contagion+"/"+sRatio+"/BoxPlot.png"
    fig.savefig(figpath, dpi=300)
```

```python
for dataset in datasets:
    for contagion in contagions:
        for sRatio in startRatio:
            path=output_folder_path+dataset+"/"+contagion+"/"+sRatio
            print(path)
            create_graphs(path, dataset, contagion, sRatio)
```
