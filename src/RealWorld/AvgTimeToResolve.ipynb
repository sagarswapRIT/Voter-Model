```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
folder_path="../../data/output/RealWorld/twitchRU/ComplexContagion/05/"

files=os.listdir(folder_path)
data={}
average={}

for file in files:
    if os.path.isfile(folder_path+"/"+file):
        with open(os.path.join(folder_path, file), 'r') as f:
            content = pd.read_csv(folder_path+"/"+file, names=['Epoch', 'Pop', 'Frac', 'DiscEdge'], skiprows=1, sep=" ")
            file=file.split('_')
            r=file[1]
            finalEpoch=content.iloc[-1, 0]
            if r in data:
                data[r].append(finalEpoch)
            else:
                data[r]=[]
                data[r].append(finalEpoch)
for k, v in data.items():
    avg=sum(v)/len(v)
    average[k]=avg

print(average)
```

```python
fig=plt.figure(figsize=(10, 6))
ax = fig.subplots()
for k, v in average.items():
    k=int(k[1:])/100
    print(k)
    ax.scatter(k, v, color='black', s=300)
plt.title('Average time taken to resolve twitchRU with Complex Contagion for Static Network')
plt.xlabel("Rewiring Rate ($\\alpha$)")
plt.ylabel("Time", fontsize='large')
plt.show()
```
